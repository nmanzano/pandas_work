```python
#MOVIES REVIEWED BY THE NY TIMES DURING THE MONTH OF NOVEMBER 2018
import pandas as pd
import requests
from pandas.io.json import json_normalize

# imported the The New York Times API
url = "https://api.nytimes.com/svc/movies/v2/reviews/search.json"

# Included unique API Key I signed up for
headers = {'api-key': "483085154d4d47ba893e48ec30de9dca"}

# Converted into JSON
response = requests.get(url, headers=headers).json()

# Converted JSON into a table readable by Pandas
df = json_normalize(response['results'])

# Removed Columns that were not necessary
df = df.drop(['critics_pick', 'date_updated', 'link.suggested_link_text', 'link.type', 'link.url', 'multimedia.height', 'multimedia.src', 'multimedia.type', 'multimedia.width'], axis=1)

# Renamed Column Names for better understanding
df.rename(columns={'byline':'Critic Name',
                   'display_title':'Movie Title',
                   'headline':'Critic Headline',
                   'mpaa_rating': 'MPAA Rating',
                   'opening_date': 'Release Date',
                   'publication_date': 'Review Publication Date',
                   'summary_short': 'Movie Summary'
                  }, 
                 inplace=True)

# Replaced Empty cells that did not have rating and replaced it with 'Not Yet Rated'
df['MPAA Rating'] = df['MPAA Rating'].replace('', 'Not Yet Rated')

# Removed NaN in Release Date with 'Not Yet Released'
df['Release Date'] = df['Release Date'].fillna('Not Yet Released')

df
```
