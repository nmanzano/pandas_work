```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Structured data and imported file
#Converted floats to numbers
#renemaed column to person

movie_data = pd.read_csv('movie_data.csv').fillna(0)

movie_data.iloc[:,1:] = movie_data.iloc[:,1:].astype('int64')
movie_data.rename(columns = {'Unnamed: 0':'Person'}, inplace = True)
movie_data
```

```python
#Calculated the average for each user
#Calculated the average for each movie
#Reset the index and renamed row to Average 4 movies
#Rounded each number to whole number and converted to whole number.

movie_data = pd.DataFrame(movie_data)
col = movie_data.loc[:]
movie_data['Average 4 User'] = col.mean(axis=1)

add_movie_avg = movie_data.append(movie_data.mean().rename('Average 4 Movies'))

final = add_movie_avg.fillna('Average 4 Movies').reset_index(drop=True)
final.iloc[:,1:] = np.around(final.iloc[:,1:]).astype('int64')
final
```

```python
#Created Graph representation 

final.plot.area();
```
