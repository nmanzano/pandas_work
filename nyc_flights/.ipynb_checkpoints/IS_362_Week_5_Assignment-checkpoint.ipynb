```python
# step 1) Import the weather file
# step 2) Select the column 'time_hour'
# step 3) Seeking only the data and not time, I slice the values to display date.
# step 4) Assign the values from step 3 to the column 'time_hour' for the main table
# step 5) Search for the rows that have the date February 12th, 2013
# step 6) I search the weather_df table with the results of step 5, sort the values by wind_speed in descending order. Then show top 5 results

# step 1)
weather_df = pd.read_csv('nycflights13/data-raw/weather.csv')

# step 2)
time_column = weather_df['time_hour']

# step 3)
only_date = time_column.str.slice(0, 10)

# step 4)
weather_df['time_hour'] = only_date

# step 5)
find_date = weather_df.time_hour == '2013-02-12'

# step 6)
results = weather_df[find_date].sort_values('wind_speed', ascending=False).head()

results
```
