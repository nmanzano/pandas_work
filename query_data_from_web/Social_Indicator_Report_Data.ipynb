```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#I imported the tables and after researching it I decided to tidy it up by selecting specific information

Social_Indicator = pd.read_csv('Social_Indicator_Report_Data.csv')
Social_Indicator
```

```python
#I selected from the Dimension Category the Race/Ethnicity and the years from 2014-2017
#I renamed Dimension Category to Race/Ethnicity
#I reset the index of the rows
#Leaving me with the Enrolled Full Day Pre-K Students Borough

column_year = Social_Indicator[['Dimension', '2014', '2015', '2016', '2017']].rename(columns={'Dimension': 'Race/Ethnicity'})

results = column_year.iloc[6:10].reset_index(drop=True)
results
```

```python
#Deciding to graph this information, I converted the percentage into whole numbers

results['2014'] = (results['2014'].str.rstrip('%').astype('float') / 100.0) * 100
results['2015'] = (results['2015'].str.rstrip('%').astype('float') / 100.0) * 100
results['2016'] = (results['2016'].str.rstrip('%').astype('float') / 100.0) * 100
results['2017'] = (results['2017'].str.rstrip('%').astype('float') / 100.0) * 100

results
```

```python
#I created a bar chart which shows the growth of each Race/Ethnicity from 2014-2017
graph = results.plot.bar(title='Enrolled Full Day Pre-K Students Borough')
graph.set_xticklabels(["Asian/Pacific Islander", "Black non-Hispanic", "Hispanic", "White non-Hispanic"])
```
